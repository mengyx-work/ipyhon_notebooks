#### process the predictions with the target together

```python
num_class = 4
```

```python
def softmax_array(array):
    return np.exp(array) / np.sum(np.exp(array), axis=0)

def convert_raw_result(preds, num_class=5):
    class_softmax_list = [[] for _ in xrange(num_class)]
    for pred in preds:
        softmax_res = softmax_array(pred)
        for i, res in enumerate(softmax_res):
            class_softmax_list[i].append(res)
    return class_softmax_list
    
def generate_class_name(class_num):
    class_name = "class_{}".format(class_num)
    return class_name
```

```python
results = convert_raw_result(preds)
```

```python
content_dict = {'target_class': data.target}
```

```python
for i in xrange(num_class):
    class_name = generate_class_name(i)
    content_dict[class_name] = results[i]
combined_data = pd.DataFrame(content_dict)
```

```python
results = []
for selected_class in xrange(num_class):
    class_name = generate_class_name(selected_class)
    class_boolean_target = [1 if e == selected_class else 0 for e in combined_data['target_class']]
    print 'class name: ', class_name
    class_preds = combined_data[class_name].values.tolist()
    precision, recall, threshold = precision_recall_curve(class_boolean_target, class_preds)                                                      
    res = {'target': class_boolean_target, 'pred': class_preds, 'precision': precision, 'recall': recall}
    results.append(res)
```
